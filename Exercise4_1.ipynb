## Part 1

Create a function in the cell below to convert temperatures from Fahrenheit to Celsius.

- The function should be called `fahr_to_celsius`
- It should have one parameter called `temp_fahrenheit`
- It should return a single value, the converted temperature
- You should include some inline comments

````python
# Function to convert Fahrenheit to Clesius
def fahr_to_celsius(temp_fahrenheit):
    """
    Convert temperature from °F to °C

    Parameter:
      temp_fahrenheit = temperatures in °F (float)

    Return:
      temperature converted to °C (float)
    """
    celsius = (temp_fahrenheit - 32) / 1.8  # Conversion formula from °F to °C
    return celsius
````

**Use** your newly created function to print answers to the following questions.

````python
# 1. What is 48° Fahrenheit in Celsius?
print(f"48° Fahrenheit is {fahr_to_celsius(48):.2f} in Celsius ")  # Convert 48° Fahrenheit to Celsius, 2 decimal spaces

# 2. What about 71° Fahrenheit in Celsius?
print(f"71° Fahrenheit is {fahr_to_celsius(71):.2f} in Celsius ")  # Convert 71° Fahrenheit to Celsius, 2 decimal spaces
````
